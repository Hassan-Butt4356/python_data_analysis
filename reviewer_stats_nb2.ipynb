```python
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
from openpyxl.formatting.rule import ColorScaleRule
from openpyxl.styles import Color

def auto_adjust_column_widths(file_path):
    """
    Automatically adjusts the column widths of all worksheets in an Excel workbook.
    
    Parameters:
    - file_path: The path to the Excel workbook file.
    """
    # Load the workbook
    workbook = load_workbook(file_path)
    
    # Iterate through each worksheet in the workbook
    for worksheet in workbook.worksheets:
        # Iterate through columns in the worksheet
        for column_cells in worksheet.columns:
            # Find the maximum length of content in each column
            length = max(len(str(cell.value)) if cell.value is not None else 0 for cell in column_cells)
            adjusted_width = (length + 2) * 1.2  # Adjust width for aesthetics
            
            # Set the column width
            column_letter = get_column_letter(column_cells[0].column)
            worksheet.column_dimensions[column_letter].width = adjusted_width
    
    # Save the workbook with adjusted column widths
    workbook.save(file_path)

def apply_color_scales_by_header(file_path, red_headers, green_headers):
    """
    Applies color scale formatting to specified columns across all worksheets in an Excel workbook.

    Parameters:
    - file_path: Path to the Excel workbook.
    - red_headers: List of headers for columns that should have a red-to-white color scale.
    - green_headers: List of headers for columns that should have a green-to-white color scale.
    """
    # Load the workbook
    workbook = load_workbook(file_path)
    
    # Iterate through each worksheet
    for sheet in workbook.worksheets:
        # Find column indices for red and green headers
        rows = list(sheet.iter_rows(values_only=True))
        if not rows:  # Skip sheets that are empty
            continue

        header_row = rows[0]#header_row = next(sheet.iter_rows(values_only=True))
        red_columns = [get_column_letter(header_row.index(header) + 1) for header in red_headers if header in header_row]
        green_columns = [get_column_letter(header_row.index(header) + 1) for header in green_headers if header in header_row]

        # Apply red-to-white color scale
        for column in red_columns:
            red_rule = ColorScaleRule(start_type='min', start_color=Color('#FA8072'),
                                      end_type='max', end_color=Color('FFFFFF'))
            sheet.conditional_formatting.add(f'{column}2:{column}{sheet.max_row}', red_rule)

        # Apply green-to-white color scale
        for column in green_columns:
            green_rule = ColorScaleRule(start_type='min', start_color=Color('98FB98'),
                                        end_type='max', end_color=Color('FFFFFF'))
            sheet.conditional_formatting.add(f'{column}2:{column}{sheet.max_row}', green_rule)

    # Save the workbook with the applied color scales
    workbook.save(file_path)
upper_municipality = municipality.upper()
```
