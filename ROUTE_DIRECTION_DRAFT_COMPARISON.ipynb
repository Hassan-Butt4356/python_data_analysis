```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
draft_df=pd.read_excel('od_20241203_buffalo_ny_draftfinal.xlsx')
draft_df.head()
```

```python
df=pd.read_csv('ROUTE_DIRECTION_REFACTOR_ROUTE_SPECIFIC002.csv')
df.head()
```

```python
draft_df.columns
```

```python
df.columns
```

```python
df.rename(columns={'id':'ID'},inplace=True)
```

```python
new_df=pd.merge(draft_df[['ID', 'ROUTE_DIRECTION[Code]', 'ROUTE_DIRECTION','STOP_OFF [CLNTID]','STOP_ON [CLNTID]']],df[['ID','STOP_OFF_CLNTID',
       'STOP_ON_CLNTID', 'STOP_ON_CLINTID_NEW',
       'STOP_OFF_CLINTID_NEW',
        'ROUTE_SURVEYEDCode_New',
       'ROUTE_SURVEYED_NEW']],on='ID')
# new_df=pd.merge(draft_df[['ID', 'ROUTE_DIRECTION[Code]', 'ROUTE_DIRECTION','STOP_OFF [ADDR]','STOP_OFF [CLNTID]']],df[['ID','ROUTE_SURVEYEDCode_New','ROUTE_SURVEYED_NEW','STOP_OFF_ADDR','STOP_OFF_CLNTID','STOP_OFF_CLNTID_new','STOP_OFF_ADDR_new']],on='ID')
```

```python
new_df.sort_values(by='ID',ascending=True,inplace=True)
```

```python
new_df.head()
```

```python
new_df.columns
```

```python
# Adding the 'Same Direction' column based on the condition
new_df['Same Direction'] = new_df['ROUTE_DIRECTION[Code]'] == new_df['ROUTE_SURVEYEDCode_New']

# Converting True/False to 'True'/'False' as strings if needed
new_df['Same Direction'] = new_df['Same Direction'].map({True: 'True', False: 'False'})
```

```python
new_df.head()
```

```python
new_df.to_csv('Route Direction Comparison_ROUTE_SPECIFIC002.csv',index=False)
```
