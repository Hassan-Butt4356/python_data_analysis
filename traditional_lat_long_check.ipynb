```python
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("reviewtool_20240625_BART_TraditionalTransferFlags.csv")
```

```python
db=pd.read_csv('elvisbartca2024interceptNEW_main_export_odbc.csv')
```

```python
new_df=pd.merge(df,db[['id','ORIGIN_ADDRESS_LAT','ORIGIN_ADDRESS_LONG','DESTIN_ADDRESS_LAT','DESTIN_ADDRESS_LONG','HOME_ADDRESS_LAT','HOME_ADDRESS_LONG']],on='id',how='left')
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
df_cleaned = new_df.dropna(subset=['ORIGIN_ADDRESS_LAT', 'ORIGIN_ADDRESS_LONG', 'DESTIN_ADDRESS_LAT', 'DESTIN_ADDRESS_LONG'], how='all')
```

```python
df_cleaned.head()
```

```python
df_cleaned.shape
```

```python
# Step 2: Drop rows where missing Origin or Destination lat/long match Home lat/long
condition = (
    (df_cleaned['ORIGIN_ADDRESS_LAT'].isna() & df_cleaned['ORIGIN_ADDRESS_LONG'].isna() &
     (df_cleaned['DESTIN_ADDRESS_LAT'] == df_cleaned['HOME_ADDRESS_LAT']) & 
     (df_cleaned['DESTIN_ADDRESS_LONG'] == df_cleaned['HOME_ADDRESS_LONG'])) |
    (df_cleaned['DESTIN_ADDRESS_LAT'].isna() & df_cleaned['DESTIN_ADDRESS_LONG'].isna() &
     (df_cleaned['ORIGIN_ADDRESS_LAT'] == df_cleaned['HOME_ADDRESS_LAT']) & 
     (df_cleaned['ORIGIN_ADDRESS_LONG'] == df_cleaned['HOME_ADDRESS_LONG']))
)
```

```python
df_final = df_cleaned[~condition]
```
