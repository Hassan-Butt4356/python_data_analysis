```python
import warnings
warnings.filterwarnings('ignore')
```

```python
td_df=pd.read_csv('reviewtool_20240816_SEATTLE_Distance_Transfer_Flags.csv')
od_df=pd.read_csv('reviewtool_20240815_SEATTLE_OD_Distance_Checks.csv')
```

```python
td_df.columns
```

```python
od_df.columns
```

```python
new_df=pd.merge(od_df[['id','Final_Direction_Code','O_D_Distance_Flag_Description']],
                td_df[['id','ROUTE_SURVEYEDCode','Transfer_Distance_Flag_Description']],on='id',how='outer')
```

```python
new_df
```

```python
for _,row in new_df.iterrows():
    if not pd.isna(row['Final_Direction_Code']):
        new_df.loc[row.name,'Route_Code']=row['Final_Direction_Code']
    elif not pd.isna(row['ROUTE_SURVEYEDCode']):
        new_df.loc[row.name,'Route_Code']=row['ROUTE_SURVEYEDCode']
    else:
        new_df.loc[row.name,'Route_Code']=pd.NA
    if not pd.isna(row['O_D_Distance_Flag_Description']) and not pd.isna(row['Transfer_Distance_Flag_Description']):
        new_df.loc[row.name,'Flag Description']=row['O_D_Distance_Flag_Description'] +' '+ row['Transfer_Distance_Flag_Description']
    elif not pd.isna(row['O_D_Distance_Flag_Description']):
        new_df.loc[row.name,'Flag Description']=row['O_D_Distance_Flag_Description']
    elif not pd.isna(row['Transfer_Distance_Flag_Description']):
        new_df.loc[row.name,'Flag Description']= row['Transfer_Distance_Flag_Description']
    else:
        new_df.loc[row.name,'Flag Description']=' '
```
