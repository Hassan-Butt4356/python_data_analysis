```python
import folium

# Create a map centered at a specific location
m = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

# Add markers for origin, destination, boarding, alighting, and transfer points
points = {
    "Origin": [40.7128, -74.0060],
    "Destination": [34.0522, -118.2437],
    "Boarding": [40.7306, -73.9352],
    "Alighting": [34.0522, -118.2437],
    "Transfer": [41.8781, -87.6298]
}

for label, point in points.items():
    folium.Marker(location=point, popup=label).add_to(m)

# Add lines connecting the points to show routes
folium.PolyLine(locations=[points["Origin"], points["Boarding"]], color='blue').add_to(m)
folium.PolyLine(locations=[points["Boarding"], points["Alighting"]], color='green').add_to(m)
folium.PolyLine(locations=[points["Alighting"], points["Destination"]], color='red').add_to(m)

# Save the map to an HTML file
m.save("map.html")
```

```python
# Create a map centered at a specific location
m = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

# Add markers for origin, destination, boarding, alighting, and transfer points with custom icons
points = {
    "Origin": [40.7128, -74.0060],
    "Destination": [34.0522, -118.2437],
    "Boarding": [40.7306, -73.9352],
    "Alighting": [34.0522, -118.2437],
    "Transfer": [41.8781, -87.6298]
}

# Define custom icons
icon_origin = folium.Icon(color='blue', icon='cloud')
icon_destination = folium.Icon(color='red', icon='cloud')
icon_boarding = folium.Icon(color='green', icon='cloud')
icon_alighting = folium.Icon(color='orange', icon='cloud')
icon_transfer = folium.Icon(color='purple', icon='cloud')

for label, point in points.items():
    if label == "Origin":
        folium.Marker(location=point, popup=label, icon=icon_origin).add_to(m)
    elif label == "Destination":
        folium.Marker(location=point, popup=label, icon=icon_destination).add_to(m)
    elif label == "Boarding":
        folium.Marker(location=point, popup=label, icon=icon_boarding).add_to(m)
    elif label == "Alighting":
        folium.Marker(location=point, popup=label, icon=icon_alighting).add_to(m)
    elif label == "Transfer":
        folium.Marker(location=point, popup=label, icon=icon_transfer).add_to(m)

# Add lines connecting the points to show routes
folium.PolyLine(locations=[points["Origin"], points["Boarding"]], color='blue').add_to(m)
folium.PolyLine(locations=[points["Boarding"], points["Alighting"]], color='green').add_to(m)
folium.PolyLine(locations=[points["Alighting"], points["Destination"]], color='red').add_to(m)

# Save the map to an HTML file
m.save("map_custom_markers.html")
```
